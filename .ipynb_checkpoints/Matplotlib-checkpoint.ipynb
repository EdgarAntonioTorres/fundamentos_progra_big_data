```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
plt.plot([1,2,3,4]) # Ingresamos datos
plt.ylabel("Algunos valores") # Etiquetamos eje y
plt.show() # Visualizar
```

### Formato a los ploteos

```python
plt.plot([1,2,3,4], [1,4,9,16], 'ro') # r- red o-dots
plt.axis((0,6,0,20)) # delimitamos el plot a x{0,6}, y{0,20}
plt.show()
```

```python
# Creamos intervalos de 200ms
t = np.arange(0.,5.,0.2)

# Lineas rojas, cuadrados azules y triangulos verdes
plt.plot(t,t,'r--',t,t**2,'bs',t,t**3,'g^')
plt.show()
```

```python
data = { 'a': np.arange(50),
       'c' : np.random.randint(0,50,50),
       'd' : np.random.randn(50)} ## Diccionario
data['b'] = data['a'] + 10 * np.random.randn(50)
data['d'] = np.abs(data['d']) * 100

## x,y = 'a','b'
plt.scatter('a','b',c = 'c', s = 'd', data = data) # c = color, s = tamao
plt.xlabel('Entrada a')
plt.ylabel('Entrada b')
plt.show()
```

## Datos Categoricos

Son aquellos que pueden tomar valores fijos

```python
names = ['grupo_a','grupo_c','grupo_b']
values = [1,10,100]

plt.figure(figsize = (9,3))
plt.subplot(131)
plt.bar(names,values)
plt.subplot(132)
plt.scatter(names,values)
plt.subplot(133)
plt.plot(names,values,linewidth = 5.0)
plt.suptitle("Ploteos categoricos")
plt.show
```

## Agregando y modificando texto a plots

```python
def f(t):
    return np.exp(-t) * np.cos(2*np.pi*t)

t1 = np.arange(0.0,5.0,0.1)
t2 = np.arange(0.0,5.0,0.02)

plt.figure()
plt.subplot(211) # Filas, columnas, elemento
plt.plot (t1,f(t1),'bo',t2,f(t2),'k')

plt.subplot(212)
plt.plot(t2,np.cos(2*np.pi*t2), 'r--')
plt.show
```

```python
mu, sigma = 100,15
x = mu + sigma * np.random.rand(10000)

## Creamos un histograma
n, bins, patches = plt.hist(x,50,density = True, facecolor = 'g',alpha = 0.75)

plt.xlabel("Inteligencia")
plt.ylabel("Probabilidad")
plt.title("Histograma del IQ")
plt.text(60, 0.025, r'$\mu=100, \ \sigma=15$')
plt.axis([40,160,0,0.03])
plt.grid(True)
plt.show()
```

```python
plt.figure()
plt.subplot(211) # Filas, columnas, elemento
plt.plot (t1,f(t1),'bo',t2,f(t2),'k')
plt.annotate('mximo local', xy=(0,1), xytext =(1,.5),
            arrowprops = dict(facecolor = 'black', shrink = 0.05))

plt.subplot(212)
plt.plot(t2,np.cos(2*np.pi*t2), 'r--')
plt.show
```
